# take a sentence plot dependency tree

```python
def get_Date_attribute( string ):
    print( string )
    doc = model( string )

    print( [ token for token in doc.ents] )
    displacy.render(doc, style='dep', jupyter=True)
    
    print( [ token for token in doc if token.ent_type_ == 'DATE'] )
```

```python
# import random

sentence = data[ random.randint(0, len(data) ) ]
get_Date_attribute( sentence )
```

```python
def is_father( word, root ):
    if word == root:
        return True
    
    while word != word.head:
        word = word.head
        if word == root:
            return True
    
    return False
```
